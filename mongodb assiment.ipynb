```python
Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios is it preferred to use MongoDB over SQL databases?
#ans
MongoDB is a popular NoSQL, document-oriented database designed to handle large amounts of unstructured or semi-structured data. Unlike traditional SQL databases, which use structured tables with rows and columns, MongoDB uses JSON-like documents (BSON) that allow for flexible and hierarchical data structures.

Non-relational databases (NoSQL) are databases that do not use the traditional relational model. They are designed to handle various types of data and are often used when:

Scalability: Horizontal scaling is needed to handle large volumes of data.
Flexibility: The data model can evolve over time without affecting existing data.
Variety of Data: The data includes semi-structured or unstructured data (e.g., JSON).
Scenarios where MongoDB is preferred over SQL databases:

Flexible Schema: When the data schema is dynamic or evolves frequently.
High Volume Data: When dealing with large volumes of data and requiring horizontal scaling.
Unstructured Data: When storing complex and unstructured data types such as JSON documents.
Real-time Analytics: When performing real-time analytics or aggregations on large dataset


Q2. State and Explain the features of MongoDB.
#Ans
Document-Oriented Storage: Stores data in JSON-like documents (BSON), which allows for a flexible schema and hierarchical data storage.
Scalability: Provides horizontal scaling through sharding, enabling the distribution of data across multiple servers.
Indexing: Supports various types of indexes to improve query performance, including single-field, compound, geospatial, and text indexes.
Aggregation Framework: Offers powerful aggregation operations for querying and transforming data.
Replication: Ensures high availability through replica sets, which provide automated failover and data redundancy.
Flexible Schema: Allows documents within the same collection to have different structures, enabling easy schema evolution.
Rich Query Language: Provides a rich query language with support for CRUD operations, data manipulation, and complex queries.



Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Create a database
db = client['my_database']

# Create a collection
collection = db['my_collection']

print("Database and collection created successfully.")



Q4. Using the database and the collection created in question number 3, 
write a code to insert one record, and insert many records. Use the find() and find_one() methods to print the inserted record.

#ans
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
db = client['my_database']
collection = db['my_collection']

# Insert one record
collection.insert_one({"name": "John Doe", "age": 30, "city": "New York"})

# Insert many records
collection.insert_many([
    {"name": "Jane Smith", "age": 25, "city": "Los Angeles"},
    {"name": "Mike Johnson", "age": 40, "city": "Chicago"}
])

# Print the inserted records
print("Single record:")
print(collection.find_one({"name": "John Doe"}))

print("All records:")
for record in collection.find():
    print(record)





Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

#Ans

from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
db = client['my_database']
collection = db['my_collection']

# Query documents
results = collection.find({"age": {"$gt": 30}})

# Print results
for document in results:
    print(document)


Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
#Ans

from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Access the database and collection
db = client['my_database']
collection = db['my_collection']

# Sort documents by age in ascending order
sorted_results = collection.find().sort("age", 1)

# Print sorted results
for document in sorted_results:
    print(document)


Q7. Explain why delete_one(), delete_many(), and drop() are used.

#Ans
    delete_one(): Deletes a single document that matches the specified query.


collection.delete_one({"name": "John Doe"})
Explanation: Useful for removing a specific document based on a query.

delete_many(): Deletes multiple documents that match the specified query.


collection.delete_many({"age": {"$lt": 30}})
Explanation: Useful for removing multiple documents that meet a certain condition.

drop(): Deletes an entire collection from the database.


collection.drop()
```
